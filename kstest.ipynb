```
import numpy as np
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt
```

```
x=stats.norm.rvs(size=1000)
sns.set_style('whitegrid')
sns.kdeplot(np.array(x))
plt.show()
```

```
stats.kstest(x,'norm')
```

```
y=np.random.uniform(0,1,10000)
sns.kdeplot(np.array(y))
plt.show()
```

```
stats.kstest(y,'norm')
```
